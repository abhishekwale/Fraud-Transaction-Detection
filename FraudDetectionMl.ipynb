```python
#lets import basic libraray
import pandas as pd
import numpy as np
```

```python
#reading the dataset
data = pd.read_csv('Fraud.csv')
data.head()
```

```python
#checking the end of dataset
data.tail()
```

```python
data.shape
```

```python
#check the presence of missing data
data.isna().sum()
```

```python
#total no. of fraud transac
data.isFraud.value_counts()
```

```python
#removing the unnecessary data part from dataset
data=data.drop(['nameOrig','nameDest'],axis=1)
```

```python
# Import label encoder
from sklearn import preprocessing
```

```python
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
```

```python
#grouping the fraud and non frauds seperately
data['type']= label_encoder.fit_transform(data['type'])
X, y = data.loc[:, data.columns != 'isFraud'], data['isFraud']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.40,random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
#Create a Gaussian Classifier
gnb = GaussianNB()
```

```python
#Train the model using the training sets
gnb.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = gnb.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
print ("Accuracy : ", metrics.accuracy_score(y_test, y_pred))
```
